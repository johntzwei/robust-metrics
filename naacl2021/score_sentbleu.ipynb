```python
from nltk.translate.bleu_score import sentence_bleu
from tqdm import tqdm
```

```python
references = list(open('references.txt'))
scores = []
for r, o in tqdm(zip(references, open('outputs.txt')), total=len(references)):
    r, o = r.strip(), o.strip()
    scores.append(sentence_bleu([r], o))
```

```python
scores
```

```python
f = open('sentbleu_scores', 'wt')
for line in scores:
    f.write(str(line) + '\n')
f.close()
```
