# Analysis

```python
sns.scatterplot(x='score', y='bleurt',
                data=wmt19_toen.sample(200))
```

### Confidence intervals of system-level BLEURT scores

```python
for i, group in wmt19_toen.groupby('lp'):
    if i != 'kk-en':
        continue
    
    means = group.groupby('system').mean()
    corr = means.corr()['score']['bleurt']
    
    plt.figure()
    plt.title('%s, r=%.2f' % (i, corr))
    sns.scatterplot(x='score', y='bleurt', data=means)
    
    for name, row in means.iterrows():
        if name == 'DBMS-KU-KKEN.6726':
            plt.annotate('POI', (row['score'], row['bleurt']))
```

```python
one_system = wmt19_toen[wmt19_toen.system == 'DBMS-KU-KKEN.6726']

trials = []
for i in range(0, 1000):
    sample = one_system.sample(len(one_system), replace=True)
    means = sample.mean()
    trials.append(means['bleurt'])
sns.distplot(trials)
```

### MSE across segments

```python
n = 5
for lp in wmt19_toen.lp.unique():
    wmt19_toen_lp = wmt19_toen[wmt19_toen.lp == lp].copy()
    wmt19_toen_lp['bin'] = pd.qcut(wmt19_toen_lp['score'], n, labels=range(1, n+1))
    
    mse_s = []
    for i, group in wmt19_toen_lp.groupby('bin'):
        mse = np.sum(np.square(group['score'] - group['bleurt']))
        mse_s.append(mse)
        
    print(mse_s)
    plt.figure()
    plt.title(lp)
    plt.ylabel('MSE')
    plt.xlabel('bin')
    ax = sns.scatterplot(x=range(1, n+1), y=mse_s)
```
