```python
import pickle
import pandas as pd
```

```python
ra_bert_score = pickle.load(open('../pickles/randomized_approx_bert_score.pkl', 'rb'))
ra_bleurt = pickle.load(open('../pickles/randomized_approx_bleurt.pkl', 'rb'))
```

```python
bad_bert = ra_bert_score[(ra_bert_score.human_decision != 'NS') & (ra_bert_score.human_decision != ra_bert_score.metric_decision)]
bad_bert
```

```python
bad_bleurt = ra_bleurt[(ra_bleurt.human_decision != 'NS') & (ra_bleurt.human_decision != ra_bleurt.metric_decision)]
bad_bleurt
```

```python
print(len(bad_bert))
print(len(bad_bleurt))
```
